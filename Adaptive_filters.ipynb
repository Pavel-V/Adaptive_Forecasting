```python
from pandas import Series
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
np.random.seed(0)

amplitude = 0
t = 500
random.seed()
times = np.arange(t)
noise = np.random.randn(t)
signal = 15 * np.sin(times * 0.025) + times / 10

signal_noise = signal + 3 * noise

signal_filtered_1 = np.zeros(t)
a_1 = 0.02
signal_filtered_1[0] = signal_noise[0]

signal_filtered_2_1 = np.zeros(t)
signal_filtered_2_2 = np.zeros(t)
signal_filtered_2_1[0] = signal_noise[0]
signal_filtered_2_2[0] = 0
a_2_1 = 0.1
a_2_2 = 0.25

for i in range(1, t):
    signal_filtered_1[i] = a_1 * signal_noise[i] + (1.0 - a_1) * signal_filtered_1[i - 1] 

    signal_filtered_2_1[i] = a_2_1 * signal_noise[i]+ (1.0 - a_2_1) * (signal_filtered_2_1[i - 1] + signal_filtered_2_2[i - 1])
    signal_filtered_2_2[i] = a_2_2 * (signal_filtered_2_1[i] - signal_filtered_2_1[i - 1]) + (1.0 - a_2_2) * signal_filtered_2_2[i - 1]

    
plt.plot(times, signal, color='g', linewidth = 2)
plt.plot(times, signal_noise, color='r')
plt.plot(times, signal_filtered_1, color='b', linewidth = 2)
plt.plot(times, signal_filtered_2_1, color='c', linewidth = 2)
#plt.plot(times, signal_filtered_2_2, color='m', linewidth = 2)

plt.legend(['signal', 'signal + noise', 'signal+noise filtered'])

plt.show() 

plt.rcParams["figure.figsize"] = [15, 10]
```
